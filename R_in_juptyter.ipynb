The set of all possible vectors that you can reach with a linear combination of a given pair ov vectors is called the span of those vectors. The span of $\vec{v}$ and $\vec{w}$ is the set of all their linear combiations: $a \vec{v} + b \vec{w}$. Let $a$ and $b$ vary over all $\mathbb{R}$

In a vector $\vec {v}$ with xy coordinate system, there are two very special vectors. 
$$
\vec {v} = \begin{bmatrix} 3 \\ -2 \end{bmatrix} 
$$

The unit vector in the x direction ($\hat{i}$) and the unit vector in the y direction ($\hat{j}$).  The vectors that these coordinates describe is the sum of two scaled unit vectors:

$$
(3)\hat{i} + (-2)\hat{j} \\
a\hat{v} + b\hat{w}
$$

$\hat{i}$ and $\hat{j}$ are the basis vectors of the xy coordinate system and $a$ and $b$ are linear combination of the vectors. When one of the vectors has the same span as the the other vector we say that they are linearity dependent. 

$$
\vec{u} = a \vec{v} + b \vec{w}
$$ 

For some values of $a$ and $b$. On the other hand, the vectors are "linearly independent" when:

$$
\vec{u} \ne a \vec{v} + b \vec{w}
$$

For all values of $a$ and $b$.

Support vector regression is a generalization of the classification (SVM) for continuous values. 
- SVM solves binary classification 
- SVM uses complex optimization problem to find the maximum margin

Decision boundary is in the middle of the binary classification and SVM wants to separate the data optimizyng the hyperplane. The distance of the decision boundary is determined by the support vectors. <br>
SVR uses $\varepsilon$-insensitive region also called $\varepsilon$-tube with a regression line in the center and the goal is:
- Define the $\varepsilon$-insensitive loss
- Find most of the training instances

It is important to reformulate SVM. SVM is very useful and flexible what makes superior to simple linear regression.

Decision boundary ($y_i$) is defined by the inner product of $w$ and $x_1$ plus the slope of the line $b$ plus the $\varepsilon$-deviation
$$
y_i = \langle w, x_i \rangle + b + \varepsilon \\
y_i = \langle w, x_i \rangle + b - \varepsilon \tag{2}
$$



SVM tries to find a line/hyperplane (in multidimensional space) that separates two classes of data. Then it classifies the new point depending on whether it lies on the positive or negative side of the hyperplane depending on the classes predict. <br>
A **kernel** helps us to find a hyperplane in the higher dimensional space without increasing the computational cost.
- The computational cost will increase if the dimension of the data increases
- The increase in dimension is required when we are unable to find a separating hyperplane in a given dimension and required to move in a higher dimension

A **hyperplane** is a separating line between two data classes in SVM. In **SVR** this is the line that will be used to predict the continuous output. <br>
The **Decision boundary** can be thought as a demarcation line on each side of which lie positive or negative samples (See formula 2). The new samples can be classified as either positive or negative. The same concept applies in SVM and SVR. 

The problem is SVR is ti find a function that approximates mapping from an input domain to ral numbers on the basis of a training sample. The desition boundary lines are draw at a distance $+\varepsilon$ and $-\varepsilon$ from the hyperplane. The equation for the hyperplane is:

$$
Y = wx + b
$$

The the equations of decision boundary become:

$$
wx +b = +\varepsilon \\
wx +b = -\varepsilon
$$

Any hyperplane in SVR should satisfy:

$$
-\varepsilon < Y - wx + b < +\varepsilon
$$

The train function in caret is the train function. ```train()``` evaluates by resampling, the effect of the model tuning parameters on performance. Choose the "optimal" model across these parameters. Estimates model performance from a training set. ```trainControl()``` function is used to specify the type of resampling.
- cv k-fold cross-validation (once or repeated using ```repeatedcv```)
- LOOCV leave-one-out cross-validation
- boot boostrap (simple estimation or the 632 rule using ```boot```)
When you include ```savePredictions = 'final'``` the function will saves predictions for optimal tuning parameter.

````r
train(target ~ ., # target in function of predictors (.) all predictors
      data = train, # your data
      method = "gbm", # your method
      trControl = trainControl(method = "cv") # your resampling method
      metric = "kappa" # changes the evaluate performance metric
     )

set.seed(123) # ensures that the same resamples are used between calls to train

fitControl <- trainControl(method = "repeatedCV", number = 10, repeats = 3, 
                           savePredictions = "final", sampling = "down")
# in this example I am repeating three separated times using 10-fold cv each time.
# The subsampling can be down, up, smote or rose
fitControl <- trainControl(method = "repeatedCV", number = 10, repeats = 3, savePredictions = "final",
                           allowParallel = T, summaryFunction = "twoClassSummary")
# summaryFunction compute performance metrics
# allowParallel logical to use parallel
````

```tumeGrid()``` allows the user to specify the tuning parameter grid.

````r
gbmGrid <- expand.grid(interaction.depth = c(1,5,9), n.trees = (1:30)*50, shrinkage = 0.1, n.minobsinnone = 20)
train(target ~ ., data = train, method = "gbm", 
      trControl = trainControl(method = "cv"), 
      metric = "kappa", verbose = F, tunGrid = gbmGrid)
````
